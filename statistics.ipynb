##### Quantiles [q1(25%),q2(50%),q3(75%)]

```python
import random 
x=[]
for i in range(50):
    v=random.randint(20,150)
    x.append(v)
print(x)
```

```python
import random 
x=[]
for i in range(50):
    v=random.randint(20,150)
    x.append(v)
q1,q2,q3=stats.quantiles(x)
print(q1,q2,q3)
```

#### outliers(Anomalies) 

```python
x[0]=500
x[10]=-400
print(x)
```

```python
q1,q2,q3=stats.quantiles(x)
print(q1,q2,q3)
iqr=q3-q1
print(iqr)
for i in x:
    if i<(q1-1.5*iqr):
        print(i,"low outlier")
    elif i>(q3+1.5*iqr):
        print(i,"high outlier")
```

##### standard deviation(std)

```python
x=[40,60,80]
y=[59,60,61]
z=[60,60,60]
print(stats.mean(x),stats.stdev(x))
print(stats.mean(y),stats.stdev(y))
print(stats.mean(z),stats.stdev(z))
```
